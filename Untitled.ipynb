Считаем что хромосомы одной длины, тогда просто создаем одномерный массив, а потом делаем reshape

```python
data = []
bioSamples = []#для всех образцов одинаковый, нет смысла засовывать в матрицу с образцами
for index, row in bigWig.iterrows():
    bioSamples.append(row['Biosample term name'])
    print("Download " + row['File download URL'])
    bw = pyBigWig.open(row['File download URL'])
    i = 0
    chr21Len = bw.chroms()['chr21']
    while i < chr21Len:
        j = i + binLength
        if(j > chr21Len):
            j = chr21Len

        #Если бин включен или пересекает blacklist, то не добавляем его
        #Можем потерять модификации возле стоп зон? Если они и есть, то нужны ли они для нашей задачи?
        inBlacklist = False
        for start, end in zones:
            if((j >= start) & (i <= end)):
                inBlacklist = True
                break

        if not inBlacklist:
            stat = bw.stats('chr21', i, j, exact=True)[0]
            data.append([stat, (i, j)])
            inBlacklist = False

        i = i + binStep
```

```python
bioSamples
```

```python
bins = np.reshape(np.array(data, dtype=np.object), (len(bioSamples), len(data)/len(bioSamples), 2))
```

```python
controlData = []
for index, row in metaControl.iterrows():
    print("Download " + row['Control url'])
    bw = pyBigWig.open(row['Control url'])
    i = 0
    chr21Len = bw.chroms()['chr21']
    while i < chr21Len:
        j = i + binLength
        if(j > chr21Len):
            j = chr21Len

        #Если бин включен или пересекает blacklist, то не добавляем его
        #Можем потерять модификации возле стоп зон? Если они и есть, то нужны ли они для нашей задачи?
        inBlacklist = False
        for start, end in zones:
            if((j >= start) & (i <= end)):
                inBlacklist = True
                break

        if not inBlacklist:
            controlData.append([bw.stats('chr21', i, j, exact=True)[0], (i, j)])
            inBlacklist = False

        i = i + binStep
```

```python
controlBins = np.reshape(np.array(controlData, dtype=np.object), (len(bioSamples), len(controlData)/len(bioSamples), 2))
```

Нормализуем все образцы с инпутом

```python
#TODO: попробовать разные нормализации? как понять какая лучше?
i = 0
normBins = bins.copy()
normControlBins = controlBins.copy()
normSamplesSizes = bigWig['Size'].copy().tolist()#не целые, нормально?
for r1, r2 in zip(bigWig['Size'].tolist(), metaControl['Size'].tolist()):
    n1 = (r1 + r2)/float(2*r1)
    n2 = (r1 + r2)/float(2*r2)
    normSamplesSizes[i] = n1*r1
    normBins[i, :] = map(lambda x:[0, x[1]] if x[0] is None else [x[0]*n1, x[1]], normBins[i, :])
    normControlBins[i, :] = map(lambda x:[0, x[1]] if x[0] is None else [x[0]*n2, x[1]], normControlBins[i, :])
    #Теперь вычитаем инпут из образцов
    normBins[i, :] = [[a[0] - b[0], a[1]] if a[0] - b[0] > 0 else [0, a[1]] for a, b in zip(normBins[i, :], normControlBins[i, :])]
    i = i + 1
    
```

Находим все бины в промоторах

```python
#TODO: какую область считать промотором? Непосредственно gene upstream? Сколько bp? Возьмем 1kb
#Как работает H3K27ac? не снимает же ДНК с нуклеосомы, иначе как бы мы нашли их, но тогда как позволяет транскрибироваться гену?

#А в gff3 нету случайно координат промоторов? и энхансеров? их бы тоже учесть

#Для каждого бина ищем ген 
#TODO: можно обойтись без циклов?
#https://www.biostars.org/p/172344/ - "The start coordinate of forward strand genes and the end coordinate
#of negative strand genes will represent the TSS..." - а не наоборот, ведь цепь нумеруется от 5 к 3, а полимераза идет от 3
#к 5, т.е. промотор "после" гена, если смотреть в направлении от 5 к 3
binsGenes = []
for b in normMeanBins[0]:#во всех образцах одинаковые координаты бинов
    startBinPosition = b[1][0]
    endBinPosition = b[1][1]
    
    hasGene = False
    for gene, positions in genesStartPositions.iteritems():
        if(positions[2] == 1):#+ цепь, значит TSS это start, значит делаем -1000/+1000 для получения зоны промотора
            promotorStart = positions[0] - 1000
            promotorEnd = positions[0] + 1000
        else:#- цепь, значит TSS это end, значит делаем +1000/-1000
            promotorStart = positions[1] - 1000
            promotorEnd = positions[1] + 1000
        
        #бин должен заходить в зону промотора как минимум на 150bp(почему?)
        if((endBinPosition - promotorStart >= binStep) & (promotorEnd - startBinPosition >= binStep)):
            binsGenes.append([b[1], gene])
            hasGene = True
            #бин может быть сразу в нескольких промоторах? Думаем что нет
            break;

    if(not hasGene):
        binsGenes.append([b[1], None])
    else:
        hasGene = False 
```
